```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
input_data = dataset.iloc[: , :8]
print(input_data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_data, dataset['Crop'], test_size=0.3, random_state=42)

    # Train the random forest model
model_crop = RandomForestClassifier(n_estimators=100, random_state=42)
model_crop.fit(X_train, y_train)

    # Make predictions
predicted_crop = model_crop.predict(X_test)
print(X_test)
```

```python
model_crop.predict([[4, 0,	40,	45,	25,	6.5, 800, 20]])
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(predicted_crop, y_test)
print(acc)
```

```python

import pickle 
  
# Save the trained model as a pickle string. 
saved_model = pickle.dumps(model_crop) 
  
# Load the pickled model 
crop = pickle.loads(saved_model) 
  
# Use the loaded pickled model to make predictions 
crop.predict([[4, 0,	40,	45,	25,	6.5, 800, 20]]) 
```
